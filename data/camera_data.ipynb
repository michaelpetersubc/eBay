This page contains instructions for the data that is used in the interactive eBay application in this repository.  

The data is illustrated below in json format.  There are two types of records, those for buyers and those for sellers.  The first record below is for a seller.  It describes an auction. The auction has a number, $1$ in the case of the first record below.  The other fields are self explanatory except for fields like `bidder`.  This field (and the ones below it) are only relevant for buyers, so they are set to default values - for example `bid_amount` is set to $0.0$ in every seller record.

The records for buyers (for example the second record in the list below) start with an auction number, but provide default values for all the seller fields.  The bidder's identity, bid and bid time are all recorded.

Then, for example, you might collect all the records for a particular auction number and put them in chronological order to see the history of bidding in that auction.
